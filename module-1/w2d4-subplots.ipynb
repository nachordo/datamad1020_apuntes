# Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = np.random.random(10)
x1 = np.random.random(10)
```

# Opci√≥n 1

```python
plt.figure(figsize=(12,3)) # Dimensiones del grafico
plt.subplot(1,2,1) # (nrows, ncols, index)
plt.title("Habitaci√≥n 1")
plt.ylabel("Temperatura (K)")
plt.xlabel("Tiempo (H)")
plt.plot(x)

plt.subplot(1,2,2)
plt.title("Habitaci√≥n 2")
plt.ylabel("Presion (mmHg)")
plt.xlabel("Tiempo (d)")
plt.plot(x1)
```

# Opci√≥n 2

```python
fig,a = plt.subplots(2,2)
a[0][0].plot(x)
a[0][1].plot(x1)
a[1][0].scatter(x,x1)
a[0][0].set_title("hola")
plt.show()
```
