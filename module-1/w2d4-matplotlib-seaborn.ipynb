```python
def square(l=1,c=(0,0)):
    x=np.array([0,1,1,0,0])
    y=np.array([0,0,1,1,0])
    coord = np.array([
        [0,0],
        [0,l],
        [l,l],
        [l,0],
        [0,0]
    ]) - l/2 + c
    return coord.T
```

```python
square(5)
```

```python
plt.plot(*square(5,(74.86,-12)))
```

```python
plt.figure(figsize=(5,5))
for l in range(1,10):
    plt.plot(*square(l,(l,l)))
```

```python
def square_rot(l=1,c=(0,0),angle=0):
    # convert degrees to to radians
    angle = np.pi*angle/180
    R = np.array([
        [np.cos(angle), -np.sin(angle)],
        [np.sin(angle),np.cos(angle)]]
    )
    return (square(l,c).T @ R).T
```

```python
square_rot(5,(0,0),45)
```

```python
plt.figure(figsize=(5,5))
for angle in np.linspace(0,90,10):
    plt.plot(*square_rot(1,(0,0),angle))
```

```python
plt.figure(figsize=(5,5))
for i in np.linspace(1,21,20):
    plt.plot(*square_rot(i,(0,0),4.5*i))
```
