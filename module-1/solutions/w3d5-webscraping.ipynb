```python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.content)
```

```python
table = soup.select("tr.normal")
```

```python
table[0].select("a")[0].text
```

```python
# Using python's unicode compatibility module we can convert data into a suitable format
unicodedata.normalize("NFKC",table[0].select("a")[0].text)
```

```python
table[0].select("a")[0].text.split()
```

```python
table[0].select("td.tb_region")[0].text.strip().title()
```

```python
coord = [td.text.strip() for td in table[0].select("td")[4:8]]
coord
```

```python
lat = float(coord[0]) if coord[1] == "N" else -float(coord[0])
long = float(coord[2]) if coord[-1] == "E" else -float(coord[2])
lat,long
```

```python
df = []
for tr in table:
    table_rows = tr.select("td")
    date,time = table_rows[3].select("a")[0].text.split()
    coord = [td.text.strip() for td in table_rows[4:8]]
    lat = float(coord[0]) if coord[1] == "N" else -float(coord[0])
    long = float(coord[2]) if coord[-1] == "E" else -float(coord[2])
    row = {
        "date": date,
        "time": time,
        "region": tr.select("td.tb_region")[0].text.strip().title(),
        "lat": lat,
        "long": long
    }
    df.append(row)
```

```python
df = pd.DataFrame(df)
df
```
