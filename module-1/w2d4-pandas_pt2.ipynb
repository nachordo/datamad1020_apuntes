### Concat

```python
a = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
a = pd.DataFrame(a,columns=["a","b","c"])
a
```

```python
b = [
    ["🐼","🐍","🃏"],
    ["🤯","🧑🏻‍💻","👩🏻‍💻"],
    ["🥔","🍑","🪓"]
]
b = pd.DataFrame(b,columns=["d","e","f"])
display(a)
display(b)
```

```python
pd.concat([a,b],axis=1)
```

```python
pd.concat([a,b],axis=0,ignore_index=True)
```

```python
b_re = b.copy()
b_re.columns = a.columns
pd.concat([a,b],axis=0,ignore_index=True)
```

```python
# Much ado about nothing
pd.concat([a,b.rename(columns={k:v for k,v in zip(b.columns,a.columns)})]
          ,axis=0,ignore_index=True)
```

### Merge

```python
a = a.append([{"a":"10","b":"11","c":"12"}]).reset_index()
a["common"] = [789,26,821,404]
a
```

```python
b = b.append([{"d":"😷","e":"🧐","f":"🦷"}]).reset_index()
b["connection"] = [26,821,34,789]
b
```

```python
a.merge(b,left_on="common",right_on="connection")
```

## Different how's

```python
hows = ["left", "right", "outer", "inner"]
```

```python
for how in hows:
    print(f"HOW: {how}")
    display(a.merge(b,left_on="common",right_on="connection",how=how))
    print("-"*70)
```

- merge
    - 2 DataFrames
    - a.merge(b,left_on="",right_on"")
    - columns may have different names
- join
    - n DataFrames
    - a.join(\[b,c,d\],on="")
    - columns must have same name across all data frames
- both have the same how parameter:
    - left
    - right
    - inner
    - outer
