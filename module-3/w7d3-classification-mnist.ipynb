```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_openml
# https://www.openml.org/
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
plt.imshow(X[6].reshape(28,28), cmap="gray")
```

```python
len(X[0])
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
svc = LinearSVC()
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
conf = confusion_matrix(y_test, y_pred)
```

```python
conf = pd.DataFrame(conf)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(conf, annot=True)
```

```python
eight = [i for i,e in enumerate(y_test) if e=="8"]
```

```python
plt.imshow(X_test[eight[1]].reshape(28,28), cmap="gray")
```

```python
mistaken8 = [i for i in eight if y_pred[i]!="8"]
```

```python
len(mistaken8)
```

```python
i = 5
plt.imshow(X_test[mistaken8[i]].reshape(28,28), cmap="gray")
plt.title(y_pred[mistaken8[i]]);
```

## Gallery of shame (Mistaken 8)

```python
plt.figure(figsize=(18,20))
for i,e in enumerate(mistaken8[:110]):
    plt.subplot(10,11,i+1)
    plt.imshow(X_test[mistaken8[i]].reshape(28,28), cmap="gray")
    plt.title(y_pred[mistaken8[i]])
;
# Number 8 that our model thought that weren't
```

```python
not8 = [i for i in range(len(y_test)) if y_pred[i]=="8" and y_test[i]!="8"]
```

```python
plt.figure(figsize=(18,20))
for i,e in enumerate(not8[:300]):
    plt.subplot(30,10,i+1)
    plt.imshow(X_test[not8[i]].reshape(28,28), cmap="gray")
;
# Different numbers our model thought were 8.
```
